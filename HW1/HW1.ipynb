## 4. Pymorphy vs Mystem (2 балла)

Лемматизируйте очищенный текст с помощью mystem и pymorphy. Сравните результаты: есть ли разница, что лучше. 

Для пайморфи используйте токенизацию из mystem, чтобы исключить влияние токенизации на результат.

```python
from pymystem3 import Mystem
import os, json
from pymorphy2 import MorphAnalyzer

morph = MorphAnalyzer()
mystem = Mystem()
```

```python
analyze = mystem.analyze(clear)
```

```python
words = [parse['text'] for parse in analyze if parse.get('analysis')]
```

### Рассмотрим только несовпадающие леммы:

```python
import pandas as pd

df = pd.DataFrame(columns=["Mystem", "Pymorphy"])
arr_1 = []
arr_2 = []
arr_3 = []

for word in words:
    mystem_lemma = mystem.lemmatize(word)[0]
    pymorphy_lemma = morph.parse(word)[0].normal_form
    if mystem_lemma != pymorphy_lemma:
        arr_1.append(mystem_lemma)
        arr_2.append(pymorphy_lemma)
        arr_3.append(word)
df["Mystem"] = arr_1
df["Pymorphy"] = arr_2
df["original word"] = arr_3
pd.set_option('display.max_rows', None)
df
```

Лемматизируйте очищенный текст с помощью mystem и pymorphy. Сравните результаты: есть ли разница, что лучше.
Из 673 слов лемматизировались по-разному 86. Mystem примерно в 2 раза чаще неправильно выделял леммы, то есть, лучше использовать Pymorphy. Из анализа можно сделать некоторые выводы.

#### Основные ошибки Mystem: 
1. часто определяется неверно часть речи, отсюда и неверно выделяется лемма (например, _теплее_ в тексте это прилагательное, а не наречие)

2. иногда неправильно выделяется начальная форма слова (например, для _деньги_ начальная форма _деньга_)

3. часто выделенная майстемом лемма глагола неправильная из-за вида (например, для слова _собрал_ правильная лемма _собрать_)

#### Основные ошибки Pymorphy: 

1. часто определяется неверно часть речи, отсюда и неверно выделяется лемма (например, _данные_ в тексте это существительное, а не причастие/глагол)

2. иногда неверно лемматизируется в том смысле, что лемма слова приведена почти правильно, но не в том роде (для имен-частей речи) (например, для слова _среднего_ лемма не _среднее_, а _средний_)
